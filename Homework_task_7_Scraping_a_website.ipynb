```python
# Importing the packages that I might need

# Packages for data manipulation
import numpy as np
import pandas as pd

# Packages for web scraping
import requests
from bs4 import BeautifulSoup

# Sometimes needed for working directory
import os
```

```python
# The URL I would like to use
URL = "https://www.premierleague.com/tables"

# Requesting the HTML and extracting the content
html = requests.get(URL)
soup = BeautifulSoup(html.content,'html.parser')
```

```python
# Checking what we have
soup
```

```python
team = soup.find_all("span", class_="long")
```

```python
team
```

```python
points = soup.find_all("td", class_="points")
```

```python
points
```

```python
# Cleaning my results, extracting the text
teams_list = [i.text for i in team ]
points_list = [i.text for i in points]
```

```python
points_list
```

```python
## Creating the Dataframe
df1 = pd.DataFrame([teams_list,points_list]).T
```

```python
df1
```

```python
# Adding column names
df1.columns = ["Team", "points"]
df1
```

```python
#Saving
df1.to_csv("scraperhomework.csv")

# Finding where the file has gone
os.getcwd
```
